# 2018

```python
data_2018a = pd.read_csv("../data/RawData/2018_1.csv")
data_2018a
```

```python
data_2018b = pd.read_csv('../data/RawData/2018_2.csv')
data_2018b
```

### Exploring and organizing the data

```python
#concatenate the 2018_a and 2018_b datas

data_2018 = pd.concat([data_2018a, data_2018b], ignore_index=True)
data_2018.shape
```

```python
#checking if there are any duplicates and dropping them

data_2018.drop_duplicates(inplace=True)
```

```python
#checking the total number of null values
data_2018.isnull().sum()
```

```python
#dropping the null values if the entire row is null
data_2018.dropna(how='all', inplace=True)
```

```python
#renaming the columns
data_2018 = data_2018.rename(columns={'Cod unic candidat': 'candidate_unique_code', 'Sex': 'gender', 'Specializare': 'specialization' , 'Profil': 'profile', 'Fileira': 'field_of_study',
       'Forma de învățământ': 'form_of_education', 'Mediu candidat': 'candidate_environment', 'Unitate (SIIIR)': 'unit_SIIIR',
       'Unitate (SIRUES)': 'unit_SIRUES' , 'Clasa': 'class_year', 'Subiect ea': 'subject_ea', 'Subiect eb':'subject_eb',
       'Limba modernă': 'modern_language', 'Subiect ec': 'subject_ec', 'Subiect ed': 'subject_ed', 'Promoție': 'graduation_year', 'NOTE_RECUN_A': 'grade_recognized_a',
       'NOTE_RECUN_B': 'grade_recognized_b', 'NOTE_RECUN_C': 'grade_recognized_c', 'NOTE_RECUN_D': 'grade_recognized_d', 'NOTE_RECUN_EA': 'grade_recognized_ea',
       'NOTE_RECUN_EB': 'grade_recognized_eb', 'NOTE_RECUN_EC': 'grade_recognized_ec', 'NOTE_RECUN_ED': 'grade_recognized_ed', 'STATUS_A': 'status_a',
       'STATUS_B': 'status_b', 'STATUS_C': 'status_c', 'STATUS_D': 'status_d', 'STATUS_EA': 'status_ea', 'STATUS_EB': 'status_eb',
       'STATUS_EC': 'status_ec', 'STATUS_ED': 'status_ed', 'ITA': 'ita', 'SCRIS_ITC': 'written_itc', 'SCRIS_PMS': 'written_pms', 'ORAL_PMO': 'oral_pmo',
       'ORAL_IO': 'oral_io', 'NOTA_EA': 'score_ea', 'NOTA_EB': 'score_eb', 'NOTA_EC': 'score_ec', 'NOTA_ED': 'score_ed', 'CONTESTATIE_EA': 'appeal_ea',
       'NOTA_CONTESTATIE_EA': 'appeal_score_ea', 'CONTESTATIE_EB': 'appeal_eb', 'NOTA_CONTESTATIE_EB': 'appeal_score_eb',
       'CONTESTATIE_EC': 'appeal_ec', 'NOTA_CONTESTATIE_EC': 'appeal_score_ec', 'CONTESTATIE_ED': 'appeal_ed',
       'NOTA_CONTESTATIE_ED': 'appeal_score_ed', 'PUNCTAJ DIGITALE': 'digital_score', 'STATUS': 'status', 'Medie': 'average', 'year': 'year',
       'session': 'session'})
```

```python
data_2018.columns
```

### Cleaning the data

```python
#checking the students who passed an exam at native language
native_language_students = data_2018.loc[data_2018["subject_eb"].notna(),"candidate_unique_code"]
```

```python
#dropping the students who passed exams in matern language
data_2018 = data_2018[~data_2018["candidate_unique_code"].isin(native_language_students)].drop(columns=["subject_eb", "score_eb", "appeal_score_eb", "status_eb", "appeal_eb"])
```

```python
#dropping the columns grade_recognized...
data_2018 = data_2018.drop(columns=['grade_recognized_a', 'grade_recognized_b', 'grade_recognized_c', 'grade_recognized_d', 'grade_recognized_ea', 'grade_recognized_eb', 'grade_recognized_ec', 'grade_recognized_ed', 'unit_SIRUES'])
```

```python
#creating a new column final_grade_ea
data_2018["final_score_ea"] = np.where(
    data_2018["appeal_ea"].str.lower() == "da",   # handle case-insensitive yes/no
    data_2018["appeal_score_ea"],
    data_2018["score_ea"]
)
```

```python
#creating a new column final_grade_ec
data_2018["final_score_ec"] = np.where(
    data_2018["appeal_ec"].str.lower() == "da",   # handle case-insensitive yes/no
    data_2018["appeal_score_ec"],
    data_2018["score_ec"]
)
```

```python
#creating a new column final_grade_ed
data_2018["final_score_ed"] = np.where(
    data_2018["appeal_ed"].str.lower() == "da",   # handle case-insensitive yes/no
    data_2018["appeal_score_ed"],
    data_2018["score_ed"]
)
```

```python
#dropping the columns that are no longer useful 
data_2018 = data_2018.drop(columns=['appeal_ea', 'appeal_score_ea', 'score_ea', 'appeal_ec', 'appeal_score_ec', 'score_ec', 'appeal_ed', 'appeal_score_ed', 'score_ed'])
```

```python
data_2018.shape
```

```python
data_2018
```
