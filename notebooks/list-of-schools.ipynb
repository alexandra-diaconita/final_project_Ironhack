```python
list_schools = pd.read_excel('../data/RawData/list_schools.xls')
list_schools
```

```python
list_schools.columns
```

```python
#dropping the unwanted columns
list_schools =  list_schools.drop(columns=['Nr', 'Denumire scurtă', 'Localitate superioară', 'Statut', 'Tip unitate', 'Formă de proprietate'])
```

```python
#renaming the columns
list_schools = list_schools.rename(columns= {'Cod': 'unit_SIIIR', 'Denumire': 'unit_name', 'Localitate' : 'locality', 'Județ': 'county'})
```
